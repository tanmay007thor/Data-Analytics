```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(10)
gaussian_samples = np.random.normal(loc=0, scale=1, size=100)
poisson_samples = np.random.poisson(lam=3, size=100)

def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data):
    mean = calculate_mean(data)
    return sum((x - mean) ** 2 for x in data) / len(data)

def calculate_std_dev(data):
    return calculate_variance(data) ** 0.5

def calculate_skewness(data):
    mean = calculate_mean(data)
    std_dev = calculate_std_dev(data)
    n = len(data)
    skewness = sum((x - mean) ** 3 for x in data) * \
        n / ((n - 1) * (n - 2) * std_dev ** 3)
    return skewness

def calculate_kurtosis(data):
    mean = calculate_mean(data)
    std_dev = calculate_std_dev(data)
    n = len(data)
    kurtosis = sum((x - mean) ** 4 for x in data) * n * (n + 1) / ((n - 1) *
                                                                   (n - 2) * (n - 3) * std_dev ** 4) - 3 * (n - 1) ** 2 / ((n - 2) * (n - 3))
    return kurtosis

def print_statistics(data, distribution_name):
    mean_val = calculate_mean(data)
    std_dev = calculate_std_dev(data)
    variance = calculate_variance(data)
    kurt = calculate_kurtosis(data)
    skewness = calculate_skewness(data)
    print(f"Statistics for {distribution_name} distribution:")
    print(f"Mean: {mean_val}")
    print(f"Standard Deviation: {std_dev}")
    print(f"Variance: {variance}")
    print(f"Kurtosis: {kurt}")
    print(f"Skewness: {skewness}")

def plot_histogram(data, distribution_name, bins):
    sns.histplot(data, bins=bins, kde=True, color='skyblue')
    plt.title(f"Histogram for {distribution_name} Distribution")
    plt.xlabel("Values")
    plt.ylabel("Probability Density")
    plt.show()

print_statistics(gaussian_samples, "Gaussian")
plot_histogram(gaussian_samples, "Gaussian", bins=30)
print_statistics(poisson_samples, "Poisson")
plot_histogram(poisson_samples, "Poisson", bins=15)
```
